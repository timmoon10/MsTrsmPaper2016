# Eigensolver timing study

```python
unif_data = np.loadtxt("unif.csv", delimiter=',')
foxli_data = np.loadtxt("foxli.csv", delimiter=',')

plt.clf()
fig, axes = plt.subplots(nrows=2, ncols=2)
plt.rc('font', family='serif')
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)
params = {
   'axes.labelsize': 8,
   'text.fontsize': 8,
   'legend.fontsize': 10,
   'xtick.labelsize': 10,
   'ytick.labelsize': 10,
   'text.usetex': False,
   'figure.figsize': [8,8]
   }
plt.rcParams.update(params)

axes[0,0].loglog(unif_data[:,0], unif_data[:,8], linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[0,0].loglog(unif_data[:,0], unif_data[:,2]+unif_data[:,3], linestyle='-', linewidth=2, marker="o", c="black", mec="black")
axes[0,0].set_title("Triangular Eigensolver Run Time")
axes[0,0].set_xlabel("$n$")
axes[0,0].set_ylabel("Run Time (s)")
axes[0,0].axis([1e3,1e5,1e-1,1e7])
axes[0,0].legend(["LAPACK (ZTREVC)", "Elemental (TriangEig)"], loc="upper left")

axes[1,0].semilogx(unif_data[:,0], unif_data[:,8]/(unif_data[:,2]+unif_data[:,3]), 'ko-', linewidth=2)
axes[1,0].set_title("Triangular Eigensolver Speedup")
axes[1,0].set_xlabel("$n$")
axes[1,0].set_ylabel("Speedup Relative to LAPACK")
axes[1,0].axis([1e3,1e5,0,80])


axes[0,1].loglog(unif_data[:,0], unif_data[:,10], linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[0,1].loglog(unif_data[:,0], unif_data[:,4], linestyle='-', linewidth=2, marker="o", c="black", mec="black")
axes[0,1].set_title("General Eigensolver Run Time")
axes[0,1].set_xlabel("$n$")
axes[0,1].set_ylabel("Run Time (s)")
axes[0,1].axis([1e3,1e5,1e0,1e7])
axes[0,1].legend = axes[0,1].legend(["LAPACK (ZGEEV)","Elemental (Eig)"], loc="upper left")

axes[1,1].semilogx(unif_data[:,0], unif_data[:,10]/unif_data[:,4], 'ko-', linewidth=2)
axes[1,1].set_title("General Eigensolver Speedup")
axes[1,1].set_xlabel("$n$")
axes[1,1].set_ylabel("Speedup Relative to LAPACK")
axes[1,1].axis([1e3,1e5,0,4])
axes[1,1].set_yticks(np.arange(0,4+1,1))

plt.show()
fig.savefig("unif.pdf", bbox_inches="tight")
```

```python
plt.clf()
fig, axes = plt.subplots(nrows=1, ncols=1)
plt.rc('font', family='serif')
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)
params = {
   'axes.labelsize': 8,
   'text.fontsize': 8,
   'legend.fontsize': 10,
   'xtick.labelsize': 10,
   'ytick.labelsize': 10,
   'text.usetex': False,
   'figure.figsize': [4,4]
   }
plt.rcParams.update(params)

el_pos = 1.3
lapack_pos = 0.3
width = 0.4
plt.bar(el_pos, unif_data[5,2]+unif_data[5,3], width, sum(unif_data[5,1:2]), color='black', edgecolor='black')
plt.bar(el_pos, unif_data[5,1], width, sum(unif_data[3,1:1]), color='white', edgecolor='black')
plt.bar(lapack_pos, unif_data[5,5], width, sum(unif_data[5,5:5]), color='white', edgecolor='black', hatch='.')
plt.bar(lapack_pos, unif_data[5,6], width, sum(unif_data[5,5:6]), color='white', edgecolor='black', hatch='///')
plt.bar(lapack_pos, unif_data[5,7], width, sum(unif_data[5,5:7]), color='white', edgecolor='black')
plt.bar(lapack_pos, unif_data[5,8], width, sum(unif_data[5,5:8]), color='black', edgecolor='black')
plt.ylabel("Run Time (s)")
plt.axis([0,2,0,80000])
plt.xticks([lapack_pos+width/2, el_pos+width/2])
axes.set_xticklabels(('LAPACK','Elemental'))
plt.legend(["Triangular Eigensolver", "Schur Decomposition"])
plt.tick_params(bottom='off', top='off', right='off')
plt.title("General Eigensolver Run Time")

plt.show()
fig.savefig("unif-parts.pdf",bbox_inches="tight")
```

```python
plt.clf()
fig, axes = plt.subplots(nrows=1, ncols=2)
plt.rc('font', family='serif')
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)
params = {
   'axes.labelsize': 8,
   'text.fontsize': 8,
   'legend.fontsize': 10,
   'xtick.labelsize': 10,
   'ytick.labelsize': 10,
   'text.usetex': False,
   'figure.figsize': [8,4]
   }
plt.rcParams.update(params)

n = np.array([1000, 2000, 5000,10000])
unif_error_el = np.array([1.20553e-14,1.47979e-14,1.98707e-14,2.44217e-14])
unif_error_lapack = np.array([1.20737e-14,1.48346e-14,1.99447e-14,2.45451e-14])
unif_cond_el = np.array([1023.92,928.538,2506.04,4221.7])
unif_cond_lapack = np.array([1023.92,928.538,2506.04,4221.7])

axes[0].loglog(n, unif_error_lapack, linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[0].loglog(n, unif_error_el, linestyle='--', linewidth=2, marker="o", c="black", mec="black")
axes[0].set_title("Eigenvalue Decomposition\n Relative Residual")
axes[0].set_xlabel("$n$")
axes[0].set_ylabel("Relative Residual")
axes[0].axis([1e3,1e4,1e-14,1e-13])
axes[0].legend(["LAPACK (ZGEEV)","Elemental (Eig)"], loc="upper left")

axes[1].loglog(n, unif_cond_lapack, linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[1].loglog(n, unif_cond_el, linestyle='--', linewidth=2, marker="o", c="black", mec="black")
axes[1].set_title("Eigenvector Matrix Condition Number")
axes[1].set_xlabel("$n$")
axes[1].set_ylabel("2-norm Condition Number")
axes[1].axis([1e3,1e4,1e2,1e4])
axes[1].legend(["LAPACK (ZGEEV)","Elemental (Eig)"], loc="upper left")

plt.show()
fig.savefig("unif_error.pdf", bbox_inches="tight")
```

# Pseudospectra timing study

```python
ps_image_times_el = np.array([0.225329, 0.420923, 5.20312, 39.5972, 160.946])
ps_image_times_eigtool = np.array([2.2257,2.8547,8.5572,67.242,259.88])
ps_image_n     = 100
ps_image_N     = np.array([100, 1024, 10000, 102400, 409600])
ps_matrix_times_el = np.array([0.720248, 2.35875, 5.20312, 12.0315, 49.8117, 326.289])
ps_matrix_times_eigtool = np.array([2.5257,4.3869,8.5572,37.162,315.59,3020.7])
ps_matrix_n     = np.array([10, 32, 100, 320, 1000, 3200])
ps_matrix_N     = 10000
ps_image_iters_el = np.array([674, 6912, 67386, 689785, 2758422])
ps_image_iters_eigtool = np.array([794,8148,79373,807575,3242365])
ps_matrix_iters_el = np.array([45713,67193,67386,67504,67353,67389])
ps_matrix_iters_eigtool = np.array([float('nan'),float('nan'),79373,78816,79183,float('nan')])
ps_image_flops_el = 2*ps_image_n*ps_image_n*ps_image_iters_el/ps_image_times_el*1e-9
ps_image_flops_eigtool = 2*ps_image_n*ps_image_n*ps_image_iters_eigtool/ps_image_times_eigtool*1e-9
ps_matrix_flops_el = 2*ps_matrix_n*ps_matrix_n*ps_matrix_iters_el/ps_matrix_times_el*1e-9
ps_matrix_flops_eigtool = 2*ps_matrix_n*ps_matrix_n*ps_matrix_iters_eigtool/ps_matrix_times_eigtool*1e-9

plt.clf()
fig, axes = plt.subplots(nrows=3, ncols=2)
plt.rc('font', family='serif')
fig.subplots_adjust(wspace=0.3)
fig.subplots_adjust(hspace=0.3)
params = {
   'axes.labelsize': 8,
   'text.fontsize': 8,
   'legend.fontsize': 10,
   'xtick.labelsize': 10,
   'ytick.labelsize': 10,
   'text.usetex': False,
   'figure.figsize': [8,12]
   }
plt.rcParams.update(params)


axes[0,0].loglog(ps_image_N, ps_image_times_eigtool, linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[0,0].loglog(ps_image_N, ps_image_times_el, linestyle='-', linewidth=2, marker="o", c="black", mec="black")
axes[0,0].set_title("$n=100$")
axes[0,0].set_xlabel("$N$")
axes[0,0].set_ylabel("Run Time (s)")
axes[0,0].axis([1e2,1e6,1e-2,1e5])
axes[0,0].legend(["EigTool", "Elemental (SpectralWindow)"], loc="upper left")

axes[1,0].semilogx(ps_image_N, ps_image_times_eigtool/ps_image_times_el, 'ko-', linewidth=2)
axes[1,0].set_title("$n=100$")
axes[1,0].set_xlabel("$N$")
axes[1,0].set_ylabel("Speedup Relative to EigTool")
axes[1,0].axis([1e2,1e6,0,10])

axes[0,1].loglog(ps_matrix_n, ps_matrix_times_eigtool, linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[0,1].loglog(ps_matrix_n, ps_matrix_times_el, linestyle='-', linewidth=2, marker="o", c="black", mec="black")
axes[0,1].set_title("$N=10000$")
axes[0,1].set_xlabel("$n$")
axes[0,1].set_ylabel("Run Time (s)")
axes[0,1].axis([1e1,1e4,1e-2,1e5])
axes[0,1].legend(["EigTool", "Elemental (SpectralWindow)"], loc="upper left")

axes[1,1].semilogx(ps_matrix_n, ps_matrix_times_eigtool/ps_matrix_times_el, 'ko-', linewidth=2)
axes[1,1].set_title("$N=10000$")
axes[1,1].set_xlabel("$n$")
axes[1,1].set_ylabel("Speedup Relative to EigTool")
axes[1,1].axis([1e1,1e4,0,10])

axes[2,0].semilogx(ps_image_N, ps_image_flops_eigtool, linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[2,0].semilogx(ps_image_N, ps_image_flops_el, linestyle='-', linewidth=2, marker="o", c="black", mec="black")
axes[2,0].set_title("$n=100$")
axes[2,0].set_xlabel("$N$")
axes[2,0].set_ylabel("GFLOPS")
axes[2,0].axis([1e2,1e6,0,0.5])
axes[2,0].legend(["EigTool", "Elemental (SpectralWindow)"], loc="upper left")

axes[2,1].semilogx(ps_matrix_n, ps_matrix_flops_eigtool, linestyle='-', linewidth=2, marker="s", c="grey", mec="grey")
axes[2,1].semilogx(ps_matrix_n, ps_matrix_flops_el, linestyle='-', linewidth=2, marker="o", c="black", mec="black")
axes[2,1].set_title("$N=10000$")
axes[2,1].set_xlabel("$n$")
axes[2,1].set_ylabel("GFLOPS")
axes[2,1].axis([1e1,1e4,0,6])
axes[2,1].legend(["EigTool", "Elemental (SpectralWindow)"], loc="upper left")

plt.show()
fig.savefig("ps_time.pdf", bbox_inches="tight")
```
